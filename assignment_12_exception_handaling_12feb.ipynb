```python
#what is an exception in python ? write the difference between exception and 
#syntax
'''
In Python, an exception is an event that occurs during the execution of a 
program that disrupts the normal flow of the program. When an exception occurs, 
it creates an object that represents the error or exceptional condition. This 
object contains information about the type of exception and any additional 
details related to the error.

Exceptions can occur due to various reasons, such as invalid input, division 
by zero, file not found, or network errors. They can be raised explicitly using 
the raise statement or automatically by the Python interpreter when it encounters
an error condition.

Exception: 
An exception occurs during the runtime of a program when a specific 
error condition or exceptional situation is encountered. It may be caused by 
factors such as invalid input, division by zero, or accessing an unavailable 
resource. Exceptions can be handled and recovered from using exception handling 
mechanisms like try and except.

Syntax Error: 
A syntax error, also known as a parsing error, occurs when the Python 
interpreter encounters a statement or expression that violates the rules of the 
Python language grammar. Syntax errors are typically caused by mistakes in the 
code's structure, such as incorrect indentation, missing or misplaced punctuation,
or improper use of keywords or operators. These errors prevent the program from 
being executed at all and need to be fixed before the code can be run successfully.

exceptions occur during the runtime of a program due to specific error 
conditions, while syntax errors are detected by the Python interpreter during the
parsing phase and prevent the program from running. Exceptions can be handled and
recovered from using exception handling techniques, whereas syntax errors 
require code correction to resolve them.
'''

try:
    num = int(input("Enter a number: "))
    result = 10 / num
    print("Result:", result)
except ValueError:
    print("Invalid input. Please enter a valid number.")
except ZeroDivisionError:
    print("Cannot divide by zero.")
```

```python
# Q3 which python statements are used to catch and handle exceptions? with example
#In Python, the try and except statements are used to catch and handle exceptions
#. The try block contains the code that may potentially raise an exception, and 
#the except block is used to define the actions to be taken when a specific 
#exception occurs.

try:
    num1 = int(input("Enter the first number: "))
    num2 = int(input("Enter the second number: "))
    result = num1 / num2
    print("Result:", result)
except ValueError:
    print("Invalid input. Please enter a valid integer.")
except ZeroDivisionError:
    print("Cannot divide by zero.")
    
#Using try and except allows you to catch specific exceptions and handle them 
#gracefully. If an exception occurs within the try block, the program flow will 
#immediately jump to the corresponding except block, and the remaining code in 
#the try block will be skipped. This way, you can control how your program 
#handles exceptions and take appropriate actions based on the type of exception 
#encountered.





```

```python
#explain with example 
#1) try except 
#2) raise 
#3) finally

'''
try-except:
The try-except block is used to catch and handle exceptions. It allows you to 
specify the code that may raise an exception within the try block and define how
to handle that exception within the corresponding except block.
python
'''
try:
    num1 = int(input("Enter a number: "))
    num2 = int(input("Enter another number: "))
    result = num1 / num2
    print("Result:", result)
except ValueError:
    print("Invalid input. Please enter a valid integer.")
except ZeroDivisionError:
    print("Cannot divide by zero.")
'''
In this example, the try block attempts to perform division between two numbers 
provided by the user. If a ValueError exception occurs (e.g., if the user enters
a non-numeric value), the program will jump to the corresponding except 
ValueError block and display an error message. If a ZeroDivisionError occurs 
(e.g., if the user enters zero as the second number), the program will jump to 
the corresponding except ZeroDivisionError block and display another error 
message. The use of try-except allows you to handle different types of 
exceptions separately and provide appropriate feedback or alternative actions.

'''
'''
raise:
The raise statement is used to explicitly raise an exception in Python. It 
allows you to create and raise your own custom exceptions or propagate built-in 
exceptions.
'''
def validate_age(age):
    if age < 0:
        raise ValueError("Age cannot be negative.")
    elif age < 18:
        raise ValueError("You must be 18 years or older.")

try:
    user_age = int(input("Enter your age: "))
    validate_age(user_age)
    print("Access granted.")
except ValueError as ve:
    print("Error:", str(ve))
'''  
In this example, a validate_age() function is defined to check if a given age 
is valid. If the age is less than zero, a custom ValueError exception is raised
with a specific error message. If the age is less than 18, another custom 
ValueError exception is raised. In the try block, the user is prompted to enter 
their age, and the validate_age() function is called. If an exception is raised 
within the function, it will be caught in the except ValueError block, and the 
error message will be displayed.
'''
'''
finally:
The finally block is used to specify code that will be executed regardless of whether an exception was raised or not. It is generally used to perform cleanup operations, such as closing files or releasing resources, that need to be done regardless of the exception handling.
python
'''
try:
    file = open("data.txt", "r")
    # Perform some file operations
except FileNotFoundError:
    print("File not found.")
finally:
    file.close()
'''
In this example, the try block attempts to open a file named "data.txt" for 
reading and perform some operations on it. If a FileNotFoundError exception 
occurs (e.g., the file does not exist), the program will jump to the 
corresponding except FileNotFoundError block and display an error message. 
Regardless of whether an exception occurs or not, the finally block will be 
executed, ensuring that the file is closed properly using the file.close() 
statement.

The finally block is useful when you need to ensure that certain actions are 
always performed, such as closing files, releasing resources, or cleaning up 
after executing critical code, irrespective of exceptions occurring.'''
```

```python
'''
Custom exceptions in Python are user-defined exceptions that allow you to create your own exception classes to handle specific error conditions in your code. These exceptions extend the base Exception class or any of its subclasses and provide additional functionality or context-specific behavior.

Here are some reasons why we need custom exceptions:

Specificity: Custom exceptions allow you to define more specific error conditions that are relevant to your code or domain. By creating custom exceptions, you can provide more descriptive error messages and handle specific exceptional cases in a more meaningful way.

Modularity and Reusability: Custom exceptions make your code more modular and reusable. By defining specific exceptions for different scenarios, you can easily handle those exceptions in a consistent manner across multiple parts of your codebase. It promotes code organization and helps in code maintenance and debugging.

Clarity and Readability: Custom exceptions improve code clarity and readability by making it easier to understand the intention and purpose of exception handling. When you encounter a custom exception, it gives you a clear indication of the specific error condition or exceptional situation that occurred.
'''
class InvalidUsernameException(Exception):
    pass

def validate_username(username):
    if len(username) < 5:
        raise InvalidUsernameException("Username must be at least 5 characters long.")

try:
    user_input = input("Enter a username: ")
    validate_username(user_input)
    print("Username is valid.")
except InvalidUsernameException as e:
    print("Invalid username:", str(e))


    
'''
In this example, a custom exception called InvalidUsernameException is defined by creating a new class that extends the base Exception class. The validate_username() function checks if a given username is valid and raises the InvalidUsernameException if it is less than 5 characters long.

In the try block, the user is prompted to enter a username, and the validate_username() function is called. If the username is invalid, an InvalidUsernameException is raised. This exception is caught in the except InvalidUsernameException block, and an appropriate error message is displayed.

By using a custom exception, we can specifically handle the case of an invalid username in a more organized and readable manner. It also allows us to differentiate it from other exceptions and provide more specific feedback to the user.





'''
```
